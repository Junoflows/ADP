```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np

data = load_diabetes()
X = data['data']
y = data['target']
```

```python
df = pd.DataFrame(np.c_[X, y], columns = data['feature_names'] + ['target'])
df.head()
```

```python
import statsmodels.api as sm

X = df['bmi']
y = df['target']

X = sm.add_constant(X)
model = sm.OLS(y, X)
fitted = model.fit()
resid = fitted.resid
print(fitted.summary())
```

```python
# sklearn 방법
from sklearn.linear_model import LinearRegression

X = df[['bmi']]
y = df[['target']]

lr = LinearRegression()
lr.fit(X, y)

y_pred = lr.predict(X)

print(lr.coef_)
print(lr.intercept_)
print(lr.score(X,y))
```
