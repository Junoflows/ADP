#### 1-2 이상치를 제거하는 방법을 서술하고, 이상치 제거 후 결과를 통계적으로 나타내시오.

이상치를 탐지하는 방법으로 boxplot 을 그려 시각적으로 확인하는 방법이 있고, 수치적으로 확인하는 방법으로는 IQR 방식을 이용하는 방법이 있다.  
IQR 방식이란 사분위수를 기준으로 Q3-Q1 를 IQR 로 설정하고 Q3+1.5*IQR 보다 크거나, Q1-1.5*IQR 보다 작으면 이상치로 간주하는 방법이다.

```python
# boxplot 을 이용한 이상치 시각화
## 수치형 데이터가 아닌 Marital_Status 와 수치적인 의미가 없는 ID 를 제외하고 box plot을 그리자
data.head(3)

cols = data.columns.drop(['ID', 'Marital_Status'])
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 5, figsize = (15, 10))
for i, col in enumerate(cols):
    ax[i // 5, i % 5].boxplot(data[col])

plt.show()
```

```python
cols = data.columns.drop(['ID', 'Marital_Status'])
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(3, 5, figsize = (15, 10))
plt.subplots_adjust(wspace=0.5)

for i, col in enumerate(cols):
    sns.boxplot(data[col], ax = ax[i // 5, i % 5])
    ax[i // 5, i % 5].set_ylabel(col)

plt.show()
```

+ boxplot 으로 시각화해보니 Income 에 이상치가 있음을 확인하였다.  
+ IQR 방식으로 이상치 데이터 값을 살펴보자.

```python
q3 = data['Income'].quantile(0.75)
q1 = data['Income'].quantile(0.25)
IQR = q3 - q1
data[(data['Income'] > q3 + 1.5*IQR) | (data['Income'] < q1 - 1.5*IQR)]['Income']
```

666666 을 제외한 7개의 수치는 특정 고소득층으로 해석할 수 있으므로 666666 만 평균으로 대체해주기로 하자.

```python
data.loc[2233, 'Income'] = np.nan
data['Income'] = data['Income'].fillna(data['Income'].mean())
```

```python
sns.boxplot(data['Income'])
plt.show()
```

#### 1-3 위에서 전처리한 데이터로 Kmeans, DBSCAN 등의 방법으로 군집을 생성하시오.

```python
data1 = pd.get_dummies(data, columns=['Marital_Status'], drop_first= True).astype('float')
data1

# Kmeans 군집분석을 하기 위해 엘보우 기법으로 최적의 군집 개수를 찾아보자.
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings(action='ignore')

def elbow(X):
  sse = []
  for i in range(1, 11):
    km = KMeans(n_clusters = i, random_state = 42)
    km.fit(X)
    sse.append(km.inertia_)

  plt.plot(range(1,11), sse, marker = 'o')
  plt.xlabel('The number of clusters')
  plt.ylabel('SSE')
  plt.show()

elbow(data1)
```

엘보우 방법으로 찾은 최적의 군집 수는 3개로 보인다.

```python
kmeans = KMeans(n_clusters= 3, random_state= 42)
kmeans.fit(data1)
```

```python
data1['cluster'] = kmeans.labels_
data1
```

```python
from sklearn.cluster import DBSCAN

data2 = data1.copy()
dbscan = DBSCAN(eps = 0.5, min_samples=5).fit(data2.iloc[:, :-1])
data2['cluster'] = dbscan.labels_

data2.cluster.value_counts()

# DBSCAN 으로는 군집이 잘 되지 않음을 알 수 있다.
```

### 3. A,B,C 자치구별 W 의원에 대한 찬성, 반대 지지를 나타낸다. 자치구별 지지율이 같은지에 대해서 검정하라
![image](https://github.com/Datamanim/datarepo/blob/main/adp/26/problem3.png?raw=true)

```python
data = pd.DataFrame({'A' : [176, 124], 'B' : [193, 107], 'C' : [159, 141]}, index = ['찬성', '반대'])

from scipy.stats import chi2_contingency
# chi2_contingency(data)

statistic, pvalue, df, excepted = chi2_contingency(data)
print('statistic :', statistic)
print('pvalue :', pvalue)
print('df :', df)
print('excepted :', excepted)
```

독립성 검정을 시행하자.  
+ H0 : 두 변수가 독립이다.
+ H1 : 두 변수가 독립이 아니다.
p-value 가 0.05보다 작으므로 귀무가설을 기각한다. 즉 지지율과 지역이 관계가 연관이 있다.

### 4. A학교 남여 학생들의 평균 혈압 차이가 있는지 여부에 대한 검정하시오.
단, 남학생과 여학생의 혈압 데이터는 정규분포를 따르며 등분산임을 가정

#### 4-1 남녀 학생들의 평균 혈압 차이가 있는지에 대해 가설을 설정하시오.

```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/26_problem6.csv')
data
```

+ H0 : 남녀 학생들의 평균 혈압이 같다.
+ H1 : 남녀 학생들의 평균 혈압이 다르다.

#### 4-2 검정 통계량을 구하고 판단하시오.

```python
from scipy.stats import ttest_ind

data.gender = data.gender.map({'male' : 0, 'female' : 1})

male = data.loc[data.gender == 0, 'pressure']
female = data.loc[data.gender == 1, 'pressure']
```

```python
ttest_ind(male, female, equal_var= True)
```

pvalue 가 0.05보다 크므로 귀무가설을 기각할 수 없다. 즉 남녀의 혈압의 평균은 같다.

#### 4-3 평균 혈압차의 신뢰구간을 구했을 때, 판단한 결과가 (2)의 결과를 지지하는지 설명하시오.

```python
from scipy.stats import t
import numpy as np
mean_m = male['pressure'].mean()
mean_f = female['pressure'].mean()

std_m = male['pressure'].std(ddof=1)
std_f = female['pressure'].std(ddof=1)

n1, n2 = 16, 9
d = mean_m - mean_f

Sp = np.sqrt((std_m**2)/n1 + std_f**2/n2)

tstats = 1.38
ME = tstats * Sp

d + ME, d - ME
```

```python
from scipy.stats import t
import numpy as np
mean_m = male['pressure'].mean()
mean_f = female['pressure'].mean()

std_m = male['pressure'].std(ddof=1)
std_f = female['pressure'].std(ddof=1)

n1, n2 = 16, 9
d = mean_m - mean_f
conf_a = 0.05 # 신뢰수준

Sp = np.sqrt((std_m**2)/n1 + std_f**2/n2)

tstats = t.ppf(1-conf_a/2, 23)
ME = tstats * Sp

d + ME, d - ME
```

### 5. height(키), weight(몸무게), waist(허리둘레) 컬럼을 가진 problem7.csv파일을 가지고 다음을 분석하라.
A시의 20대 남성 411명을 임의로 추출한 후 키, 몸무게, 허리둘레를 조사하여 기록한 데이터이다. 이 데이터를 이용하여 20대 남성의 키와 허리둘레가 체중에 영향을 미치는지 알아보시오.

#### 5-1 아래 조건을 참고하여 회귀계수를 구하시오.
- 베이지안 회귀
- 시드넘버 1234로 지정
- 1000번의 burn-in 이후 10,000의 MCMC를 수행
- 회귀계수의 사전분포는 부적절한 균일분포(inproper uniform prior distribution), 오차항의 분산의 사전분포는 역감마 분포로 지정. 이때, 형상(Shape)모수와 척도(Scale)모수는 각각 0.005로 지정.

```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/26_problem7.csv')
data
```

```python
from sklearn.linear_model._bayes import ARDRegression, BayesianRidge
from sklearn import linear_model
help(linear_model)
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['weight'], axis = 1)
y = data.weight

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 42)

ARD_reg = ARDRegression(n_iter= 1000, alpha_2 = 0.005, lambda_1 = 0.005, fit_intercept=False)
ARD_reg.fit(X_train, y_train)

print(ARD_reg.coef_)
```
