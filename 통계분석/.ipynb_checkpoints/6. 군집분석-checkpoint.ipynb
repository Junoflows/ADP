## 6. 군집분석

### 6.1 계층적 군집분석

n 개의 군집으로 시작해 점차 군집의 개수를 줄여나가는 방법

가. 종류
+ 최단연결법
  + n X n 거리행렬에서 거리가 가장 가까운 데이터를 묶어서 군집을 형성함
+ 최장연결법
  + 데이터와의 거리를 계산할 때 최장거리를 거리로 계산함
+ 평균연결법
  + 데이터와의 거리를 계산할 때 평균을 거리로 계산함
    + 와드연결법
    + 군집 내 편차들의 제곱합을 고려한 방법
    + 군집 간 정보의 손실을 최소화하기 위해 군집화를 진행함


나. 군집화
+ 거리행렬을 통해 가장 가까운 거리의 객체들 관계를 규명함
+ 덴드로그램 그리기
+ 덴드로그램의 최상단부터 세로축의 개수에 따라 가로선을 그어 군집 개수를 선택함
+ 각 객체들의 구성을 고려해 적절한 군집수 생성함

함수1. linkage 함수
> linkage(y, method = 'single', metric = 'euclidean')

+ y : 거리행렬(데이터를 넣으면 됨)
  + 모든 요소가 유한해야함(NaN, inf가 없어야함)

* method:
    - single : 최단연결법
    - complete : 최장연결법
    - average : 평균연결법
    - centroid : 중심연결법
    - ward : 와드연결법
* metric:
    - euclidean :유클리디언 거리
    - seuclidean : 표준화거리
    - mahalanobis : 마할라노비스 거리
    - chebyshev : 체비셰프 거리
    - cityblock : 맨하탄거리
    - canberra : 캔버라 거리
    - minkowski : 민코우스키 거리
    - jaccard : 자카드 거리
    - cosine : 코사인 유사도

함수2. dendrogram 함수

> dendrogram(Z, orientation='top', labels=None, color_threshold=None, get_leaves=True, ,    
* Z : linkage matrix. linkage 함수로 생성함
* orientation
    - top: 플롯이 위에서 아래방향으로 그려짐
    - bottom: 플롯이 아래에서 위방향으로 그려짐
    - left: 플롯이 오른쪽에서 왼쪽방향으로 그려짐
    - right: 플롯이 왼쪽에서 오른쪽방향으로 그려짐

### 6.2 계층적 군집분석 실습

USArrests 데이터의 정보로 거리를 구하고 최단, 최장, 평균연결법을 실시해보아라

이 데이터 세트에는 1973년에 미국 50개 주 각각에서 폭행, 살인 및 강간으로 체포된 인구 100,000명당 통계가 포함되어 있습니다. 또한 도시 지역에 거주하는 인구 비율도 제공됩니다.

4개의 변수에 대한 50개의 관측값이 있는 데이터 프레임.

* Murder: 살인 체포(100,000명당)
* Assault: 폭행 체포(100,000명당)
* UrbanPop : 도시 인구 비율
* Rape:  강간 체포(100,000명당)

```python
# plotting 환경 설정
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,5) # 그래프 크기
plt.rcParams['font.family'] = 'Malgun Gothic' # 글꼴
plt.rcParams['font.size'] = 12 # 글꼴 크기
plt.rcParams['axes.unicode_minus'] = False # 마이너스 표기
```

```python
import pandas as pd
import numpy as np

US = pd.read_csv('https://raw.githubusercontent.com/phong0104/Alex_Lee_SNU_EduTech/9e65152ae7b230d9c940d03dbdc9e468f69cf426/Certificate/(2)%20%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D%20%EC%A0%84%EB%AC%B8%EA%B0%80%20%EC%8B%A4%EA%B8%B0/Ver.2/USArrests.csv')
US.head()
```

```python
US.columns = ['State', 'Murder', 'Assault', 'UrbanPop', 'Rape']
labellist = US.State.tolist()
US.head()
```

```python
# 최단연결법
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
single = linkage(US.iloc[:, 1:], method= 'single', metric = 'euclidean')

# 덴드로그램 그리기
dendrogram(single,
           orientation='top',
           labels = labellist,
           distance_sort= 'descending',
           color_threshold=25, # 군집의 수를 설정하는 height 값 설정
           show_leaf_counts=True)
plt.axhline(y = 25, color = 'r', linewidth = 1) # height 값에 따라 선을 그어 적절한 군집 수 설정

plt.show()
```

fcluster(Z, t, criterion='distance')    
* Z: linkage 함수로 반환되는 matrix로 구성된 계층군집
* t: 상수. flat clusters 형성할 때 적용하는 임계값. 덴드로그램의 color_threshold랑 일치하게 설정하면 그 값에 따른 군집결과 확인 가능
* criterion: flat cluster 형성하는 기준
    - distance: 각 군집의 원래 관측치가 t보다 큰 동위원소 거리를 갖지 않도록 군집을 형성

```python
# state 마다 어떤 군집이 설정되었는지 확인
from scipy.cluster.hierarchy import fcluster

assignments = fcluster(single, 25, 'distance')
US['cluster'] = assignments
US.head()
```

```python
# 최장연결법
complete = linkage(US.iloc[:, 1:-1], method = 'complete', metric = 'euclidean')

# 군집간의 거리 차이가 길다 : 군집이 잘 이루어졌다.

# 덴드로그램 그리기
dendrogram(complete,
           orientation='top',
           labels = labellist,
           distance_sort = 'descending',
           color_threshold= 100)
plt.axhline(y = 100, color = 'r', linewidth = 1)
plt.show()

assignments = fcluster(complete, 100, 'distance')
US['cluster'] = assignments
```

```python
# 평균 연결법
average = linkage(US.iloc[:, 1:-1], metric = 'euclidean', method = 'average')

# 덴드로그램 그리기
dendrogram(average,
           orientation='top',
           labels = labellist,
           distance_sort='descending',
           color_threshold= 70,
           show_leaf_counts=True)

plt.axhline(y = 70, color = 'r', linewidth = 1)
plt.show()

assignments = fcluster(average, 70, 'distance')
US['cluster'] = assignments
```

```python
# 와드 연결법
ward = linkage(US.iloc[:, 1:-1], method = 'ward', metric='euclidean')

# 덴드로그램
dendrogram(ward,
           orientation= 'top',
           labels= labellist,
           distance_sort= 'descending',
           color_threshold= 250)

plt.axhline(y = 250, color = 'r', linewidth = 1)
plt.show()

assignments = fcluster(ward, 250, 'distance')
US['cluster'] = assignments
```

### 6.3 비계층적 군집분석

#### 6.3.1 K-mean
가. 개념
+ 주어진 데이터를 k 개의 클러스터로 묶는 알고리즘
+ 각 클러스터와 거리 차이의 분산을 최소화 하는 방식으로 작동
  + n개의 객체들을 각 집합 내 객체간 응집도를 최대로 하는 k개의 집합 S를 찾는 것이 목표
  + 응집도 : 각 집합별 중심점 ~ 집합 내 객체들 간 거리의 제곱합

나. 표준 알고리즘  
1. i번째 클러스터의 중심을 mu_i, 클러스터에 속하는 점의 집합을 S_i 라고 할 때, 전체 분산을 최소화하는 S를 찾는 것이 목표
2. 초기의 mu_i 를 설정
3. 클러스터 설정 : 각 데이터로부터 각 클러스터들의 mu_i 까지의 유클리드 거리를 계산해, 해당 데이터에서 가장 가까운 클러스터를 찾아 데이터를 배당함
4. 클러스터 중심 재조정 : mu_i 를 각 클러스터에 있는 데이터들의 무게중심 값으로 재설정
+ 알고리즘을 실행하는 동안 여러 개의 최근접 이웃 검색이 필요하기 때문에 이 알고리즘은 확장성이 높지 않습니다. 알고리즘은 중심 변화가 작을 때 알고리즘은 반복을 중지합니다.

최적의 군집개수를 결정하는 방법
+ 콜린스키 하라바츠 스코어
+ 엘보우 방법
+ 실루엣 방법

KMeans(n_clusters = 8, random_state = None)
+ n_clusters : 클러스터의 개수

#### 6.3.2 KMeans 실습
iris 데이터로 kmeans 군집분석을 해보자

```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')

iris = pd.read_csv('https://raw.githubusercontent.com/phong0104/Alex_Lee_SNU_EduTech/9e65152ae7b230d9c940d03dbdc9e468f69cf426/Certificate/(2)%20%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D%20%EC%A0%84%EB%AC%B8%EA%B0%80%20%EC%8B%A4%EA%B8%B0/Ver.2/data/iris.csv')
iris.head()
```

```python
X = iris.drop(['target'], axis = 1)
y = iris['target']
# 원본 데이터 iris target 시각화
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(iris,
             diag_kind= 'kde',
             hue = 'target')
plt.show()
```

```python
# Kmeans 군집화 실행
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, random_state=42)
km.fit(X)

km.labels_
```

최적의 클러스터 개수를 결정하는데 사용되는 방법 (1)
+ calinski_harabasz (콜린스키 하라바츠 지수)
  + 모든 클러스터에 대한 클러스터 간 분산과 클러스터 내 분산의 합의 비율
  + 점수가 높을수록 성능이 좋음
  + 장점
    + 계산이 빠름
  + 단점
    + 밀도 기반 클러스터에서 점수가 더 높게 나타나는 경향이 있음

+ completeness_score
  + 클러스터 내의 데이터 포인트가 얼마나 잘 모여있는지를 측정한다.
  + 1에 가까울수록 각 클러스터에 속한 데이터 포인트가 서로 비슷하고 잘 군집화되었음을 의미.

```python
# 군집분석 평가척도 : 콜린스키 하라바츠
from sklearn.metrics import calinski_harabasz_score, completeness_score
print(calinski_harabasz_score(X, km.labels_))
print(completeness_score(y, km.labels_))
```

```python
# 클러스터별 콜린스키 하라바츠 결과 비교
scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters = k, random_state=42).fit(X)
    labels = kmeans.labels_
    scores.append(calinski_harabasz_score(X, labels))

pd.DataFrame(scores, index = [x for x in range(2, 10)], columns = ['calinski_harabasz_score'])
```

+ iris 데이터는 군집이 3일 때, 가장 점수가 높음

최적의 클러스터 개수를 결정하는데 사용되는 방법 (2)
+ 엘보우 기법
  + 클러스터 내 오차제곱합을 클러스터 개수마다 비교
  + 클러스터 개수를 늘려나가면서 계산한 SSE 을 비교
  + 어느 순간 줄어드는 비율이 급격하게 작아지는 부분과 기울기가 소실되는 구간이 생김(elbow)
  + 그 부분이 최적의 클러스터 개수임

```python
import matplotlib.pyplot as plt

sse = []
for i in range(1, 11):
  km = KMeans(n_clusters = i, random_state = 42)
  km.fit(X)
  sse.append(km.inertia_)

plt.plot(range(1,11), sse, marker = 'o')
plt.xlabel('The number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
sse
```

[결과 해석]  
3에서 4로 갈 때, 기울기가 완만해지므로 군집 수는 3이 적당함


최적의 클러스터 개수를 결정하는데 사용되는 방법 (3)
+ 실루엣 계수
    + 보통 0.5 이상이면 좋은 수치라고 판단

```python
from sklearn.metrics import silhouette_samples, silhouette_score

for i in range(2,10):
    kmeans = KMeans(n_clusters= i, random_state= 42).fit(X)
    labels = kmeans.labels_
    print(silhouette_score(X, labels = labels))
```

+ 실루엣 계수로 보았을 때 2~3개의 군집으로 나누는게 효과적

```python
# 최적의 K로 kmeans 군집화 실행
km = KMeans(n_clusters = 3, random_state=42).fit(X)

# 할당된 군집을 iris 열 추가
iris['cluster'] = km.labels_
print(iris.cluster.value_counts())
print(iris.target.value_counts())
iris.head()
```

```python
# Kmeans 시각화
# 군집결과 시각화
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(iris, diag_kind='kde', hue = 'cluster')
plt.show()
```

```python
# 비계층적 군집분석
# Kmeans, DBSCAN 군집분석 최종 코드
## Kmeans
from sklearn.cluster import KMeans

X = iris.drop(['target'], axis = 1)

import matplotlib.pyplot as plt
import seaborn as sns

# 원본 데이터 시각화
print('[원본 데이터 시각화]')
sns.pairplot(iris,
             diag_kind='kde',
             hue = 'target',
             palette= 'bright')
plt.show()

# kmeans 군집화 -> 최적의 클러스터 개수 찾기
## 1. calinski_harabasz_score
scores = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i, random_state=42).fit(X)
    labels = kmeans.labels_
    scores.append(calinski_harabasz_score(X, labels))

print()
print('calinski_harabasz_score : \n',scores) # 가장 값이 큰 인덱스가 최적의 군집 -> 3

## 2. elbow method
def elbow(X):
  sse = []
  for i in range(1, 10):
    km = KMeans(n_clusters = i, random_state = 42)
    km.fit(X)
    sse.append(km.inertia_)

  plt.plot(range(1,10), sse, marker = 'o')
  plt.xlabel('The number of clusters')
  plt.ylabel('SSE')
  plt.show()
  print(sse)

print()
print('[elbow method]')
elbow(X) # 2에서 3으로 갈 때 기울기가 완만해짐 -> 최적의 군집은 3

## 3. 실루엣 계수 -> 시각화 함수를 정의해줘야 해서 시험중에는 무리일 듯
from sklearn.metrics import silhouette_score

for i in range(2,10):
    scores = []
    kmeans = KMeans(n_clusters= i, random_state= 42).fit(X)
    labels = kmeans.labels_
    scores.append(silhouette_score(X, labels = labels))

print()
print('[silhouette score] : \n', scores) # 0.5 이상이면 좋은 수치라고 판단


## 최적의 K로 kmeans 군집화 실행
kmeans = KMeans(n_clusters= 3, random_state=42).fit(X)
labels = kmeans.labels_
iris['cluster'] = labels
# print(iris.cluster.value_counts())
# print(iris.target.value_counts())

## 군집결과 시각화
print()
print('[군집결과 시각화]')
sns.pairplot(iris,
             diag_kind = 'kde',
             hue = 'cluster',
             palette = 'bright')

plt.show()
```

### 6.4 혼합 분포 군집
가. 개념
  + 모형 기반(model-based) 군집방법
  + 데이터가 k개의 모수적 모형의 가중합으로 표현되는 모집단 모형으로부터 나왔다는 가정
  + k개의 각 모형은 군집이며, 각 데이터는 추정된 k개의 모형 중 어느 모형으로부터 나왔을 확률이 높은지에 따라 군집의 분류가 만들어짐

나. 혼합분포모형으로 설명할 수 있는 데이터의 형태
+ 다봉형 등

다. 사용 알고리즘 : EM(Expectation-Maximization)
+ 각 자료에 대해 Z의 조건부분포로부터 조건부 기대값을 구할 수 있음
  + Z의 조건부분포 : 어느 집단에 속할지에 대한 확률분포
+ 관측변수 X와 잠재변수 Z를 포함하는 (X, Z)에 대한 로그-가능도함수에 Z 대신 상수값인 Z의 조건부 기댓값을 대입하면, 로그-가능도함수를 최대로 하는 모수를 쉽게 찾을 수 있음
  + 이 과정을 최대가능도추정이라고 함
+ EM 알고리즘 4단계 정리
  1. 초기값 설정 : 필요한 모수에 대해 초기값을 설정한다.
  2. E : 잠재변수 Z 의 기대치 계산(X가 특정 군집에 속할 확률을 계산함)
  3. M : 잠재변수 Z의 기대치를 이용해 파라미터를 추정함(계산된 확률을 통해 모수를 재추정함)
  4. 반복 정지 : 수렴 조건이 만족될 때까지 E와 M을 반복
    + 수렴 조건 = 최대가능도가 최대인가?

라. 혼합 분포 군집모형의 특징
+ 장점
  + kmeans 보다 좀 더 통계적으로 엄밀한 결과를 얻을 수 있음(확률분포 도입)
  + 군집을 몇 개의 모수로 표현할 수 있고, 서로 다른 크기나 모양의 군집을 찾을 수 있음
+ 단점
  + 군집의 크기가 너무 작으면 추정의 정도가 떨어짐
  + 데이터가 커지면 EM 알고리즘 적용 시 시간/계산비용 증가 -> 사용하기 어려움
  + 이상치자료에 민감하기 때문에 사전에 조치가 필요
  + 유형들의 분포가 정규분포와 차이가 크면 결과가 좋지 못함
    + 그래서 현실에 존재하는 여러 개의 확률 분포를 사용자가 혼합해서 표현함
    + 몇 개의 확률분포를 혼합할 것인지는 사용자가 지정함. (그 개수에 맞춰서 모델이 만들어짐)

#### 혼합분포 군집분석 실습
iris 데이터의 Species 를 제외하고 혼합분포군집분석을 실시하기

```python
import pandas as pd
iris = pd.read_csv('https://raw.githubusercontent.com/phong0104/Alex_Lee_SNU_EduTech/9e65152ae7b230d9c940d03dbdc9e468f69cf426/Certificate/(2)%20%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D%20%EC%A0%84%EB%AC%B8%EA%B0%80%20%EC%8B%A4%EA%B8%B0/Ver.2/data/iris.csv')

df = iris.drop(['target'], axis = 1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# 가우시안 혼합모델 구축
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components= 3)
gmm.fit(df_scaled)
gmm_labels = gmm.predict(df_scaled)
df['cluster'] = gmm_labels

# 군집 변수별 통계량 확인
print('[군집 변수별 통계량]')
print(df.groupby('cluster').mean())

# 군집결과 시각화
print()
print('[군집결과 시각화]')
import seaborn as sns
sns.pairplot(df,
             diag_kind = 'kde',
             hue = 'cluster',
             palette= 'bright')
plt.show()

print()
print('[원본 데이터 시각화]')
sns.pairplot(iris,
             diag_kind = 'kde',
             hue = 'target',
             palette= 'bright')
plt.show()
```

### 6.5 자기조직화지도 (Self-Organizing Map, SOM)
1. 개념
+ 대뇌피질과 시각피질의 학습 과정을 모델화한 인공신경망으로써 자율 학습에 의한 클러스터링을 수행하는 알고리즘이다.
+ SOM 은 우선 Map size 를 설정해줍니다. (군집분석으로 SOM을 사용할 때는 군집의 수를 차원의 수로 맞춰주어야 합니다.)
+ Default Map 을 조금씩 수정해서(학습해서 = fit) 데이터 분포 형태에 Map을 근사시킵니다.

#### 6.5.1 SOM 실습
iris 데이터를 som으로 군집분석 해보세요.

```python
# sklearn의 모든 분류기와 마찬가지로 SOM 인스턴스를 빌드하고 SOM.fit() 에 맞게 데이터 호출
# iris 데이터에 3개의 클래스가 있다는 것을 미리 알고있으므로
# 자체 구성 맵에 3 X 1 구조를 사용하지만 실제로는 데이터에 가장 적합한 것을 찾기 위해 다른 구조를 시도해봐야함

iris = pd.read_csv('https://raw.githubusercontent.com/phong0104/Alex_Lee_SNU_EduTech/9e65152ae7b230d9c940d03dbdc9e468f69cf426/Certificate/(2)%20%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D%20%EC%A0%84%EB%AC%B8%EA%B0%80%20%EC%8B%A4%EA%B8%B0/Ver.2/data/iris.csv')

X = iris.drop(['target'], axis = 1)
y = iris['target']

# scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

from sklearn_som.som import SOM
iris_som = SOM(m = 3, n = 1, dim = 4)
iris_som.fit(X_scaled, epochs= 100)

pred = iris_som.predict(X_scaled)
iris['cluster'] = pred
print('SOM 군집결과 \n', pred)

# SOM 군집결과 시각화
print()
print('[SOM 군집결과 시각화]')
sns.pairplot(iris, diag_kind='kde', hue = 'cluster', palette='bright')
plt.show()

print('[원본 데이터 시각화]')
sns.pairplot(iris, diag_kind='kde', hue = 'target', palette='bright')
plt.show()
```
