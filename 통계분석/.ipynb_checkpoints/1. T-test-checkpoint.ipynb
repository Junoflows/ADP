### 1.1 일표본 T-검정 (One Sample T-Test)

가. 일표본 T-검정 개념
+ 단일 모집단에서 관심있는 연속형 변수의 평균값을 특정 기준값과 비교
+ 가정 : 모집단의 구성요소들이 정규분포를 이룬다는 가정 - stats.shapiro()
  + 종속변수는 연속형이어야 함
  + 검증하고자 하는 기준값이 있어야 함

나. 일표본 T-검정 단계
|1단계: |가설설정|  |
|------|---|---|
|모수|모평균(μ)|  |
|귀무가설(H0)|μ=μ0|  |
|대립가설(H1)|(1) 양측검정: |μ!=μ0|
| |(2) 우단측검정: |μ>μ0|
| |(3) 좌단측검정: |μ<μ0|
      
      
|2단계: |유의수준 설정|          |
|------|---|---|
      
      
|3단계: |검정통계량의 값 및 유의확률 계산|  |
|------|---|---|
      
      
|4단계: |귀무가설의 기각여부 판단 및 의사결정|
|------|---|
|p-value:|유의확률|
|α: | 유의수준|
|p-value < α:|귀무가설을 기각하고, 대립가설을 채택|
|p-value > α:|귀무가설을 기각하지 않음|)

#### 1.1.1 일표본 T-검정 예제
cats 데이터에서 고양이들의 평균몸무게가 2.6kg 인지 아닌지 통계적 검정을 수행하고, 결과 해석하기(양측검정, 유의수준 = 0.05)

```python
import pandas as pd
cats = pd.read_csv('https://raw.githubusercontent.com/phong0104/Alex_Lee_SNU_EduTech/9e65152ae7b230d9c940d03dbdc9e468f69cf426/Certificate/(2)%20%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D%20%EC%A0%84%EB%AC%B8%EA%B0%80%20%EC%8B%A4%EA%B8%B0/Ver.2/data/cats.csv')
cats
```

```python
# cats 의 Bwt의 정규성 검정 : shapiro test
# H0 : 정규성을 가진다
# H1 : 정규성을 가지지 않는다.

import scipy.stats as stats
from scipy.stats import shapiro
mu = 2.6
shapiro(cats['Bwt'])
```

[결과 해석]
+ 검정통계량 : 0.9518
+ p-value : 6.730e-05

p-value가 유의수준 0.05보다 작으므로 귀무가설을 기각한다. 즉 정규성을 가지지 않는다.
따라서 wilcox test 로 T-검정 진행

```python
# 정규성을 만족하지 않으므로 wilcox test 실시
stats.wilcoxon(cats['Bwt']-mu, alternative= 'two-sided')
```

```python
# 정규성을 가지는 경우
from scipy.stats import ttest_1samp
ttest_1samp(cats['Bwt'], popmean= mu)
```

```python
import matplotlib.pyplot as plt

cats_Bwt_cnt = cats.Bwt.value_counts(sort = False)

width = 0.4
plt.bar(cats_Bwt_cnt.index, cats_Bwt_cnt.values)
plt.title('Bwt')
plt.xlabel('Bwt')
plt.ylabel('count')
plt.show()
```

[결과 해석]
+ 검정통계량 : 3573.0
+ p-value : 0.25524
p-value 가 유의수준 0.05보다 작으므로 귀무가설을 기각한다. 즉 고양이들의 평균 몸무게는 2.6kg 이라고 할 수 없다.

### 1.2 대응표본 T-검정 (Paried Sample T-Test)

가. 대응표본 T-검정 개념
+ __단일모집단__ 에 대해 두 번의 처리를 가했을 때, 두 개의 처리에 따른 평균의 차이를 비교
+ 표본 내의 개체들에 대해 두 번의 측정 실시
+ 가정 : 모집단의 관측값이 정규성을 만족해야 함

나. 대응표본 T-검정 단계
|1단계: |가설설정|  |
|------|---|---|
|모수|두 개 모평균 사이의 차이(D)|  |
|귀무가설(H0)|μx-μy=D=0|  |
|대립가설(H1)|(1) 양측검정: |μx-μy=D!=0|
| |(2) 우단측검정: |μx-μy=D>0|
| |(3) 좌단측검정: |μx-μy=D<0|
      
      
|2단계: |유의수준 설정|          |
|------|---|---|
      
      
|3단계: |검정통계량의 값 및 유의확률 계산|  |
|------|---|---|
      
      
|4단계: |귀무가설의 기각여부 판단 및 의사결정|
|------|---|
|p-value:|유의확률|
|α: | 유의수준|
|p-value < α:|귀무가설을 기각하고, 대립가설을 채택|
|p-value > α:|귀무가설을 기각하지 않음 --> 사전평균 = 사후평균|

#### 1.2.1 대응표본 T-검정 예제
10명의 환자 대상 수면영양제 복용 전과 후의 수면시간 측정, 영양제의 효과가 있는지를 판단하고자 함  
(표본이 정규성을 만족한다는 가정 하에 단측검정 수행, 유의수준 = 0.05)

```python
import pandas as pd
data = {'before':[7,3,4,5,2,1,6,6,5,4],
       'after':[8,4,5,6,2,3,6,8,6,5]}
data = pd.DataFrame(data)
data
```

```python
# 귀무가설 : before 평균과 after 평균이 같다.
# 대립가설 : before 평균과 after 평균이 다르다.

from scipy.stats import ttest_rel
ttest_rel(data.after, data.before, alternative='greater')
```

[결과 해석]
+ 검정 통계량 : 4.74
+ pvalue : 0.0005

pvalue 가 유의수준 0.05보다 작으므로 귀무가설을 기각한다.  
즉 수면영양제 복용 전과 후의 수면시간이 다르다라고 할 수 있고 수면시간이 늘었다는 결론

```python
# 만일 정규성 가정이 없을 때 풀이과정
from scipy.stats import shapiro
print(shapiro(data.before - data.after)) # shapiro test 에서 p-value가 0.05보다 작으므로 정규성을 만족함

# 정규성을 만족하지 않을 때
from scipy.stats import wilcoxon
wilcoxon(data.after, data.before, alternative='greater')
```

### 1.3 독립표본 T-검정(Independent Sample T-Test)

가. 독립표본 T-검정 개념
+ 두 개의 독립된 모집단의 평균을 비교
+ 모집단, 모수, 표본이 모두 2개씩 존재
+ 가정
  + 두 모집단은 정규성 만족해야함 - stats.shapiro()
  + 두 모집단은 분산이 서로 같아야 함(등분산성 가정) stats.levene()
  + 독립변수는 범주형, 종속변수는 연속형

나. 독립변수 T-검정 단계
|1단계: |가설설정|  |
|------|---|---|
|모수|서로 독립된 두 모집단의 평균(μ1, μ2)|  |
|귀무가설(H0)|μ1=μ2|  |
|대립가설(H1)|(1) 양측검정: |μ1!=μ2|
| |(2) 우단측검정: |μ1>μ2|
| |(3) 좌단측검정: |μ1<μ2|
      
|2단계: |유의수준 설정|          |
|------|---|---|
      
      
|3단계: |검정통계량의 값 및 유의확률 계산|  |
|------|---|---|
      
      
|4단계: |귀무가설의 기각여부 판단 및 의사결정|
|------|---|
|p-value:|유의확률|
|α: | 유의수준|
|p-value < α:|귀무가설을 기각하고, 대립가설을 채택|
|p-value > α:|귀무가설을 기각하지 않음|

#### 1.3.1 독립표본 T-검정 예제
성별에 따른 몸무게가 같은지 통계적 검정을 수행하고 결과 해석

```python
import pandas as pd
cats = pd.read_csv('https://raw.githubusercontent.com/phong0104/Alex_Lee_SNU_EduTech/9e65152ae7b230d9c940d03dbdc9e468f69cf426/Certificate/(2)%20%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D%20%EC%A0%84%EB%AC%B8%EA%B0%80%20%EC%8B%A4%EA%B8%B0/Ver.2/data/cats.csv')
cats
```

__정규성 검정__

```python
from scipy.stats import shapiro

male = cats[cats.Sex == 'M']
female = cats[cats.Sex == 'F']

male.shape, female.shape

# 데이터가 30개 이상이므로 정규성 가정
```

__등분산성 검정__

```python
from scipy.stats import levene
levene(male.Bwt, female.Bwt) # 등분산성을 만족하지 않음
```

__등분산성을 만족하지 않은 독립 T검정__

```python
from scipy.stats import ttest_ind
ttest_ind(male.Bwt, female.Bwt, equal_var=False)

# 평균몸무게는 다르다.
```
